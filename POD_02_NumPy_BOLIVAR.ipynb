# Operación vectorizada vs. Operación loopeada
`numpy` no es solo poderoso por la gama de operaciones que podemos hacer en pocas lineas de código. Sino que por su eficiencia.

Supongamos que tenemos la función $f(x)=10*(x^2/e^x)$. Evaluar la función (en celdas apartes) entre -1 y 1 con 1,000,000 valores (i) con y (ii) sin un loop `for`. Medir tiempos de cada celda y sacar conclusiones.

```python
#Escribir el codigo para responder la pregunta en esta celda
import time
arr3 = np.linspace(-1,1,1_000_000)
fx = np.empty(1_000_000)

start_time = time.time()  # Tiempo de inicio

# Acceder a cada elemento del array
for i in range(len(arr3)):
    fx[i] = 10 * (np.power(arr3[i],2)/np.power(np.e,arr3[i]))  # Simula procesamiento (acceso al elemento)

end_time = time.time()  # Tiempo de finalización

# Calcular el tiempo total
tiempo_total = end_time - start_time

tiempo_total
```

# Análisis de imagenes

Matematicamente hablando, las imagenes son arreglos. Si una imagen es de blanco y negro, tenemos una imagen de un canal y puede ser interpretado como una simple matriz (`.shape=2`). Si tiene varios canales, tenemos una matriz asignada para cada canal (`.shape=3`).

[Lenna](https://en.wikipedia.org/wiki/Lenna) es una imagen ampliamente utilizada en ciencias de la computación. Se volvió un icono. La idea va a ser analizar a la imagen y tratarla para varios propositos. Corra la siguiente celda para descargar la imagen y guardala como arreglo `numpy` en la variable `image`. Utilizar esta variable en las siguientes celdas.

```python
!wget https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png
from PIL import Image
image = Image.open('Lenna_(test_image).png')
image = np.asarray(image)
```

```python
#Escribir el codigo para responder la pregunta en esta celda
plt.imshow(image)
```

Recorte la imagen en ancho entre (tomando como referencia los ejes de la imagen vista anteriormente) los 100 y 350 pixeles y en alto entre 200 y 400 pixeles. Mostrar el resultado.

```python
image_np = np.asarray(image)
cropped_image_np = image_np[:250, :250]

# Convert the NumPy array back to an image
cropped_image = Image.fromarray(cropped_image_np)

plt.axis('off')
plt.imshow(cropped_image)
```

Muestre cada uno de los canales de la imagen.

```python
# Split the image into Red, Green, and Blue channels
red_channel = image_np[:, :, 0]
green_channel = image_np[:, :, 1]
blue_channel = image_np[:, :, 2]

# Plot each channel
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Display the Red channel
axs[0].imshow(red_channel, cmap='Reds')
axs[0].set_title('Red Channel')
axs[0].axis('off')

# Display the Green channel
axs[1].imshow(green_channel, cmap='Greens')
axs[1].set_title('Green Channel')
axs[1].axis('off')

# Display the Blue channel
axs[2].imshow(blue_channel, cmap='Blues')
axs[2].set_title('Blue Channel')
axs[2].axis('off')

# Show the plot
plt.show()
```

Calcule el minimo, el máximo, y el promedio de los valores de la imagen.

```python
#Escribir el codigo para responder la pregunta en esta celda
print(f"Red Channel: Min = {np.min(red_channel)}, Max = {np.max(red_channel)}, Mean = {np.mean(red_channel)}")
print(f"Green Channel: Min = {np.min(green_channel)}, Max = {np.max(green_channel)}, Mean = {np.mean(green_channel)}")
print(f"Blue Channel: Min = {np.min(blue_channel)}, Max = {np.max(blue_channel)}, Mean = {np.mean(blue_channel)}")
```

La verdad que tener todos los colores de la imagen es muy redundante. Paselo a blanco y negro. Para ello, tome el promedio de los canales en cada pixel. Muestre la imagen en blanco y negro.

```python
#Escribir el codigo para responder la pregunta en esta celda
# Calculate the grayscale image by averaging the color channels
gsimage_np = np.mean(image_np, axis=2).astype(np.uint8)

# Convert the grayscale NumPy array back to an image
gs_image = Image.fromarray(gsimage_np)

# Display the grayscale image
plt.imshow(gs_image, cmap='gray')
plt.axis('off')  # Hide axes
plt.show()
```

Por último, vamos a proceder a "binarizar" la imagen. Es decir, vamos a setear en 1 TODOS los pixeles donde la intensidad (es decir, el valor del pixel) sea mayor a 200. El resto, lo seteamos a 0. Mostrar el resultado.

```python
#Escribir el codigo para responder la pregunta en esta celda
# Binarize the image: set pixels > 200 to 1, and the rest to 0
binarized_image_np = np.where(gsimage_np > 200, 1, 0)

# Convert the binarized NumPy array back to an image
binarized_image = Image.fromarray((binarized_image_np * 255).astype(np.uint8))

# Display the binarized image
plt.imshow(binarized_image, cmap='gray')
plt.axis('off')  # Hide axes
plt.show()
```

Lea el archivo csv (sin `pandas` pero se puede usar `csv`) y quedese con todas las columnas que son numericas. Inspeccionarlo puede hacer el trabajo bastante facil.

Nota: Aunque `numpy` tenga una libreria para leer archivos, en este caso no no es útil por que todas las columas constan de diferentes tipo de dato y ademas el encabezado es `string`.

```python
import csv

# Leer el archivo CSV
with open("Boston.csv", mode='r') as file:
    csv_reader = csv.reader(file)
    headers = next(csv_reader)  # Leer los encabezados
    rows = list(csv_reader)    # Leer el resto de las filas

# Filtrar columnas numéricas (verificar con la primera fila)
numerical_indices = [
    i for i, value in enumerate(rows[0]) if value.replace('.', '', 1).isdigit()
]
numerical_headers = [headers[i] for i in numerical_indices]

# Extraer datos de las columnas numéricas
numerical_data = [
    [float(row[i]) for i in numerical_indices] for row in rows
]
```

Una vez procesado el archivo csv, proceda a convertirlo en una matriz de `numpy`.

```python
import pandas as pd

# Convertir a DataFrame de pandas
df = pd.DataFrame(numerical_data, columns=numerical_headers)

print(df)
```

Reporte el promedio de todas las columnas.

```python
# Calcular e imprimir el promedio de cada columna
for column in df.columns:
    column_mean = df[column].mean()
    print(f"Promedio de '{column}': {column_mean}")
```

Reporte la división entre una columna y otra (a elección cuales)

```python
# Dividir dos columnas (asegúrate de que las columnas existan en el archivo)
col1, col2 = numerical_headers[0], numerical_headers[1]  # Seleccionar dos columnas
df["Resultado"] = df[col1] / df[col2]

# Imprimir la nueva columna
print(f"Nueva columna 'Resultado' (división de '{col1}' y '{col2}'): \n")
print(df["Resultado"])
```
