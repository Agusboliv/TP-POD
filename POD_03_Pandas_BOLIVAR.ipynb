```python
import zipfile

with zipfile.ZipFile('flight-delays.zip', 'r') as zip_ref:
    zip_ref.extractall('flight_delays_data')
```

```python
import pandas as pd

# Lee los archivos CSV y crea DataFrames
airlines = pd.read_csv('flight_delays_data/airlines.csv')
airports = pd.read_csv('flight_delays_data/airports.csv')
flights = pd.read_csv('flight_delays_data/flights.csv', low_memory=False)

```

```python
#airlines.head()
#airports.head()
#flights.head()
```

Combine (*join*) las tablas `airlines`, `airports`, y `flights` en una sola tabla.

```python
df = pd.merge(flights, airlines, left_on='AIRLINE', right_on='IATA_CODE')

df = pd.merge(df, airports, left_on='ORIGIN_AIRPORT', right_on='IATA_CODE', suffixes=('', '_ORIGIN'))

df = df.rename(columns={'IATA_CODE': 'IATA_CODE_ORIGIN'})

df = pd.merge(df, airports, left_on='DESTINATION_AIRPORT', right_on='IATA_CODE', suffixes=('_ORIGIN', '_DEST'))

df.head()
```

Haga una imputación de datos COMPLETA del dataframe. Pueden escojer cualquier estrategia y no necesariamente todas las columnas deben seguir la misma estrategia.

```python
import numpy as np

LAT_TO_KM = 111  # 1° de latitud ≈ 111 km
LON_TO_KM_AT_EQUATOR = 111  # 1° de longitud en el ecuador ≈ 111 km

# Calcular las diferencias en grados
df["delta_lat"] = df["LATITUDE_DEST"] - df["LATITUDE_ORIGIN"]
df["delta_lon"] = df["LONGITUDE_DEST"] - df["LONGITUDE_ORIGIN"]

# Aproximar la distancia considerando la longitud ajustada por el coseno de la latitud promedio
df["distance"] = np.sqrt(
    (LAT_TO_KM * df["delta_lat"])**2 +
    (LON_TO_KM_AT_EQUATOR * np.cos(np.radians((df["LATITUDE_ORIGIN"] + df["LATITUDE_DEST"]) / 2)) * df["delta_lon"])**2
)

print(df[["LATITUDE_ORIGIN", "LONGITUDE_ORIGIN", "LATITUDE_DEST", "LONGITUDE_DEST", "distance"]]) # (km)
```
